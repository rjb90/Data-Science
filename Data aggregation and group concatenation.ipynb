```python
def peak_to_peak(series):
    return series.max() - series.min()

grouped['data1'].apply(peak_to_peak) #Esta es la manera en que puedo aplicar una funcion sobre una columna
```

### Column-wise and multiple function application

```python
!wget https://raw.githubusercontent.com/wesm/pydata-book/1st-edition/ch08/tips.csv
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

## Group-wise operations and transformations

```python
tips.groupby(['sex','smoker'])['tip'].mean() #Le digo que me agrupe por sex y smoker y que me devuelva la media de la tip
```

```python
tips.groupby(['sex','smoker'])[['total_bill','tip']].agg(['mean','std', peak_to_peak, np.count_nonzero])
```

```python
functions_list = [('range', peak_to_peak), ('number', np.count_nonzero)]

tips.groupby(['sex','smoker'])[['total_bill','tip']].agg(functions_list)
```

### Quantile and bucket analysis

```python
df2 = pd.DataFrame({
    'age':[random.randint(17, 99) for _ in range(250)],
    'salary' : [random.randint(10000, 90000) for _ in range (250)]
})

df2
```

```python
pd.cut(df2['age'], bins=(0, 18, 35, 65, 100)) #A cada aeg le asigno un subgrupo
```

```python
age_categories = pd.cut(df2['age'], bins=(0, 18, 35, 65, 100))
df2.groupby(age_categories).mean() #Y a cada subgrupo lo agrupo por la edad y le digo que me devuelva la media
```

## Pivot tables and Cross-tabulation

```python
tips
```

```python
tips.pivot_table(index=['sex', 'smoker']) #Por defeto es la media
```

Pivot table da mucha mas flexibilidad que el groupby porque pasas unos pocos parámetros y ya te lo devuelve

```python
tips.pivot_table(index=['sex','smoker'], aggfunc=np.sum)
```

```python
pivoted = tips.pivot_table(columns=['sex','smoker'], index='size' ,aggfunc=np.sum)
pivoted
```

```python
pivoted = tips.pivot_table(columns=['sex', 'smoker'], index='size',aggfunc=np.sum)
pivoted.stack()
```

```python
pivoted.unstack() #Despivoteo la tabla
```
