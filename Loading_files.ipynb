```python
import pandas as pd
import zipfile
```

# Loading CSV

Con la library zipfile puedo leer archivos csv mediante el código debajo. 

```python
zf = zipfile.ZipFile('T100_SEGMENT_ALL_CARRIER_2014.zip') #Primero le hago unzip al archivo
df = pd.read_csv (zf.open('914310910_T_T100_SEGMENT_ALL_CARRIER_2014_All.csv')) #Luego, importo el archivo que necesito
df.head()
```

```python
pd.options.display.max_columns = None
df.sample(5)
```

Una manera de explorar los datos, es usando sort_values by. De esa manera, puedo ordenar los datos por lo que quiera, hacer que me devuelva una cantidad determinada de valores y las columnas que yo quiera

```python
#sort_values me permite ordenar los datos por una columna, por defecto de menor a mayor
df.sort_values (by = 'PASSENGERS', ascending = False)[:10][['CARRIER', 'ORIGIN', 'DEST_CITY_NAME', 'PASSENGERS']]
```

```python
df['RPMs'] = df['PASSENGERS']*df['DISTANCE'] #Creo columna nueva
top_10 = df.sort_values (by = 'RPMs', ascending = False)[:10][['CARRIER','ORIGIN_CITY_NAME', 'DEST_CITY_NAME']]
top_10
```

```python
df['LF'] = df['PASSENGERS']/df['SEATS']
df.sort_values (by = 'LF', ascending = False).head(10)
```

```python
lf = df.sort_values (by = 'LF', ascending = False).head(5)['LF']
lf
```

```python
import numpy as np

lf[lf != np.inf]
```
